Popularity Based Movie Recommendation(Display 250 books with highest average rating)

```python
#merging two datatset on ISBN
ratings_with_book=ratings.merge(books,on='ISBN')
```

```python
ratings_with_book
```

### Lets find number of user ratings for each book.


```python
#Reset Index will convert the data into dataframe

num_rating=ratings_with_book.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'Num-Rating'},inplace=True)
num_rating
```

```python

num_rating_np = num_rating.values  # or num_rating.to_numpy()

print(np.sum(num_rating_np[:, 1]))
```

```python
ratings_with_book['Book-Rating'] = pd.to_numeric(ratings_with_book['Book-Rating'], errors='coerce')

avg_rating = (
    ratings_with_book.groupby('Book-Title')['Book-Rating']
    .mean()
    .reset_index()
    .rename(columns={'Book-Rating': 'Avg-Rating'})
)

avg_rating
```

```python
popular_df=num_rating.merge(avg_rating,on='Book-Title')
popular_df
```

```python
## Fetch whose rating>250 and sort it on the basis of average rating in ascending order

popular_df=popular_df[popular_df['Num-Rating']>=250].sort_values('Avg-Rating',ascending=False)
```

```python
popular_df
```

```python
# Merge
popular_df = popular_df.merge(books[['Book-Title', 'Book-Author', 'Image-URL-M']], on='Book-Title')

# Drop duplicate Book-Title rows
popular_df = popular_df.drop_duplicates('Book-Title')

# Keep only necessary columns
popular_df = popular_df[['Book-Title', 'Book-Author', 'Image-URL-M', 'Num-Rating', 'Avg-Rating']]

# View
popular_df.head()
```

```python
popular_df['Image-URL-M'][0]
```

 ### Collaborative Filtering Based recommendation
 
 ### First of all select the user id whose rating is greater than 200 and book-title greater than 50 means on which books user reviewed.

```python
x=ratings_with_book.groupby('User-ID').count()['Book-Rating']>200
exp_users=x[x].index
```

```python
#here we filter which user book rating is greater than 200 and matches the old users with the experienced users

filter_ratings=ratings_with_book[ratings_with_book['User-ID'].isin(exp_users)]
filter_ratings
```

```python
y=filter_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filter_ratings[filter_ratings['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']== pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('A Bend in the Road')
```
